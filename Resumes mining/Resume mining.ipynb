## transformation
### TF-IDF
* importance ~ appearances(term, document) / count(documents containing term in collection)
* tf = count(word) / len(document) – term frequency
* idf = log( total number of document / count(document_containing_term) )– inverse document frequency 
* tf-idf = tf * idf 

```python
#Split total term count list into small files pretending each personal resume with term count
lines_per_file = 10000
smallfile = None
with open('clean_text.txt') as bigfile:
    for lineno, line in enumerate(bigfile):
        if lineno % lines_per_file == 0:
            if smallfile:
                smallfile.close()
            small_filename = 'small_file_{}.txt'.format(lineno + lines_per_file)
            smallfile = open(small_filename, "w")
        smallfile.write(line)
    if smallfile:
        smallfile.close()
```

```python
#define tf fucntion
def computeTF(wordDict, bow):
    tfDict={}
    bowCount=len(bow)
    for word, count in wordDict.items():
        tfDict[word]=count/float(bowCount)
    return tfDict
```

```python
#define IDF funtion: 
def computeIDF(docList):
    import math
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict
```

```python
class LDA():
    def __init__(self,df_tokenized):
        self.df_tokenized=df_tokenized
        self.dictionary = Dictionary(self.df_tokenized)
        self.corpus = [self.dictionary.doc2bow(text) for text in self.df_tokenized]
```

```python
def model(self, model='LDA', workers=None, num_topics=5, update_every=5, chunksize=10000, passes=100):
        '''
        @id2word
        @num_topics
        @update_every
        @chunksize
        @passesf
        '''
        if model == 'LDA':
            self.lda = models.LdaModel(self.corpus, id2word=self.dictionary,  num_topics=num_topics, update_every=update_every, chunksize=chunksize, passes=passes)
        if model == 'LDAMultiCore':
             self.lda = models.LdaMulticore(self.corpus, workers=None, id2word=self.dictionary,  num_topics=num_topics, chunksize=chunksize, passes=passes)
        return self.lda
```

```python
def words_in_topics(self):
        if self.lda is None:
            print ('Error: please run the model function first.')
        else:
            self.w_in_t = self.lda.show_topics()
        return self.w_in_t 
```
